```python
import os
import pandas as pd
import numpy as np

# Load dataset from data folder
data_path = os.path.join('..', 'data', 'predictive_maintenance.csv')
df = pd.read_csv(data_path)

# Preview top 5 rows and column info
print("Dataset Shape:", df.shape)
df.head()
```

```python
# Clean column names: lowercase, replace spaces/brackets with underscores
df.columns = (
    df.columns.str.lower()
    .str.replace(' ', '_')
    .str.replace('[', '', regex=False)
    .str.replace(']', '', regex=False)
)

print("Cleaned Columns:\n", df.columns.tolist())
```

```python
# Check the ratio of failures vs normal operation
failure_counts = df['machine_failure'].value_counts()
failure_pct = df['machine_failure'].value_counts(normalize=True) * 100

print("--- Class Breakdown ---")
print(f"No Failure (0): {failure_counts[0]} ({failure_pct[0]:.2f}%)")
print(f"Failure    (1): {failure_counts[1]} ({failure_pct[1]:.2f}%)")
```

```python
# Select numerical sensor features and target
features_of_interest = [
    'air_temperature', 
    'process_temperature', 
    'rotational_speed', 
    'torque', 
    'tool_wear', 
    'machine_failure'
]

# Calculate Pearson Correlation matrix
corr_matrix = df[features_of_interest].corr()

# Display correlations with respect to machine_failure
print("--- Correlation with Machine Failure ---")
print(corr_matrix['machine_failure'].sort_values(ascending=False))
```
